```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import csv

# Initialize WebDriver
driver = webdriver.Chrome()

# Define XPath for each element we want to scrape
xpaths = {
    'Logo': "./div[1]//img",
    'Role': "./div[2]",
    'Company': "./div[4]/div/div[1]",
    'Location': "./div[4]/div/div[2]",
    'Source': "./div[4]/div/div[3]",
    'Posted': "./div[4]/div/div[4]/div[1]",
    'Full / Part Time': "./div[4]/div/div[4]/div[2]",
}

# Initialize data dictionary for storing scraped data
data = {key: [] for key in xpaths}

# Define the number of jobs to scrape
jobs_to_do = 100
jobs_done = 0

# Loop to scrape job listings until we collect the specified number of jobs
while jobs_done < jobs_to_do:
    try:
        # Locate all job items in the current page
        lis = driver.find_elements(By.XPATH, "//li[@data-ved]//div[@role='treeitem']/div/div")
        
        for li in lis[jobs_done:]:
            # Scroll into view to make sure the element is loaded
            driver.execute_script('arguments[0].scrollIntoView({block: "center", behavior: "smooth"});', li)
            
            # Extract data for each key (Logo, Role, Company, etc.)
            for key in xpaths:
                try:
                    # Get the text or attribute value depending on the element type
                    t = li.find_element(By.XPATH, xpaths[key]).get_attribute('src' if key == 'Logo' else 'innerText')
                except NoSuchElementException:
                    # If the element is missing, assign '*missing data*'
                    t = '*missing data*'
                data[key].append(t)
            
            jobs_done += 1
            print(f'{jobs_done=}', end='\r')  # Display jobs scraped so far
            
            # To avoid being too aggressive on requests
            time.sleep(0.2)

        # Exit the loop once the target number of jobs are scraped
        if jobs_done >= jobs_to_do:
            break

    except Exception as e:
        print(f"Error occurred: {e}")
        break

# After scraping is done, save the data into a CSV file
output_path = r'G:\DASHBORAD\portfolio_projects\Scraping\Google_Search\job_data.csv'
file_exists = os.path.isfile(output_path)

with open(output_path, mode="w", newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    if not file_exists:
        writer.writerow(xpaths.keys())  # Write header row (keys of xpaths dictionary)
    # Write the collected data for each job
    for i in range(jobs_done):
        writer.writerow([data[key][i] for key in xpaths])

print(f"Scraping complete. {jobs_done} jobs saved to {output_path}")
```

```python
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from bs4 import BeautifulSoup
import json
import time


jobs = []
total = 0


def scrape(start_url):
    """Prepare a website url and scrape all pages"""

    # Scrape all pages
    start = time.time()
    count_pages = 0
    while True:
        website_url = start_url.replace('st=0', 'st={}'.format(count_pages))
        try:
            parse(website_url)
        except TimeoutException:
            browser.quit()
            print('All data successfully scraped!')
            end = time.time()
            print('Time: {} minutes \n'.format(round((end - start) / 60), 1))
            break
        count_pages += 20

    # Store all jobs and total count
    data = {
        'total': total,
        'jobs': jobs
    }
    return json.dumps(data)


def parse(jobs_page):
    """Parse main jobs page and get all jobs URLs"""
    global total

    # Open first URL
    browser.get(jobs_page)

    # Wait or sleep till all page data loaded
    WebDriverWait(browser, 20).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, 'GXRRIBB-e-G'))
    )

    body = browser.page_source

    # Parse page html to extract jobs info
    soup = BeautifulSoup(body, 'html.parser')
    jobs_content = soup.select('.GXRRIBB-e-G')

    # Store the count of jobs located in this page
    total += len(jobs_content)

    # Get all jobs links (URLs) from jobs info (jobs_content variable)
    jobs_urls = []
    for job in jobs_content:
        job_header = job.select_one('h2 a')

        # Check that the job is related to Google company
        # To make sure that the html structure will be the same
        company = job.select_one(
            'div.sr-content div.summary .secondary-text').get_text()
        if company != 'DeepMind':
            job_link = 'https://careers.google.com/jobs' + job_header.get('href')
            jobs_urls.append(job_link)

    parse_jobs(jobs_urls)


def parse_jobs(jobs_urls):
    """Parse all jobs data"""
    global jobs

    # Open jobs URLs one by one to get page html
    jobs_html = []
    for url in jobs_urls:
        browser.execute_script("window.open('{}', 'new_window')".format(url))
        browser.switch_to.window(browser.window_handles[1])

        # Wait till all page data loaded
        try:
            WebDriverWait(browser, 20).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, 'a.secondary-text'))
            )
        except TimeoutException:
            browser.close()
            browser.switch_to.window(browser.window_handles[0])
            continue

        jobs_html.append(browser.page_source)

        # Close the current window and switch to the main window
        browser.close()
        browser.switch_to.window(browser.window_handles[0])

    # Parse all pages html to get jobs data
    jobs_list = []
    for job in jobs_html:
        soup = BeautifulSoup(job, 'html.parser')
        job_id = soup.find('div', attrs={'itemtype': 'http://schema.org/JobPosting'}).get('id')
        job_title = soup.select_one('div.card-company-job-details > h1 a.title.text').get_text()
        location = soup.select_one('div.card-company-job-details .details-panel > a').get_text()
        desc = soup.select_one('div.detail-item .description-section.text.with-benefits').get_text()
        resp_qual = soup.select('div.detail-item .description-section .GXRRIBB-S-c .description-content')
        resp = resp_qual[0].get_text()
        qual = resp_qual[1].get_text()

        job_dict = {
            'job_id': job_id,
            'title': job_title,
            'location': location,
            'intro': desc,
            'resps': resp,
            'quals': qual
        }

        jobs_list.append(job_dict)

    # Store the current jobs list with all scraped lists before
    jobs += jobs_list


if __name__ == '__main__':
    print('Start scraping all google jobs ...')

    # Prepare the URL
    base_url = 'https://careers.google.com/jobs'
    params = {
        't': 'sq',
        'li': '20',
        'st': '0',
        'jlo': 'all'
    }
    result = requests.get(url=base_url, params=params)
    url = result.url.replace('?', '#')
    
    print('Please, Do not close chrome driver. '
          'It will be closed automatically after finished.')
    print('This process may take more than 10 minutes')

    chrome_options = webdriver.ChromeOptions()
    # Set chrome in headless mode to hide the UI
    # chrome_options.add_argument('headless')
    
    # Creates and open a new instance of the chrome driver
    browser = webdriver.Chrome(options=chrome_options)

    data_json = scrape(url)
    print(data_json)
    with open('data.json', 'w') as file:
        json.dump(json.loads(data_json), file)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv

# Setup WebDriver (use Chrome or another browser of your choice)
driver = webdriver.Chrome()

# Specify the path to save the CSV file
csv_file_path = "job_listings.csv"

def scroll_page():
    """Function to scroll the page down to load more job listings."""
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  # Wait for the new content to load

try:
    # Open Google and search for jobs in New York
    search_query = "jobs in New York"
    driver.get(f"https://www.google.com/search?q={search_query}")

    # Wait for the 'Jobs' tab to be visible and click it
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//a[contains(., 'Jobs')]"))
    )
    jobs_tab = driver.find_element(By.XPATH, "//a[contains(., 'Jobs')]")
    jobs_tab.click()

    # Wait for the job listings to load initially
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, 'tNxQIb'))  # Wait for the first job listing
    )

    # Open the CSV file in write mode
    with open(csv_file_path, mode="w", newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        
        # Write header row
        writer.writerow(["Job Title", "Company Name", "Location"])
        
        # Scrape job listings and scroll the page to load more jobs
        job_titles = []
        company_names = []
        locations = []
        page_end_reached = False
        
        while not page_end_reached:
            # Extract job details using the full class names you provided (as CSS selectors)
            job_titles = driver.find_elements(By.CSS_SELECTOR, '.tNxQIb.PUpOsf')  # Full Role Class
            company_names = driver.find_elements(By.CSS_SELECTOR, '.wHYlTd.MKCbgd.a3jPc')  # Full Company Name Class
            locations = driver.find_elements(By.CSS_SELECTOR, '.wHYlTd.FqK3wc.MKCbgd')  # Full Location Class

            # Check if the number of job listings has changed after scrolling
            if len(job_titles) == 0 or len(company_names) == 0 or len(locations) == 0:
                page_end_reached = True
                break

            # Write the job details into CSV file
            for i in range(min(len(job_titles), len(company_names), len(locations))):
                writer.writerow([job_titles[i].text, company_names[i].text, locations[i].text])

            # Scroll the page to load more jobs
            scroll_page()

    print(f"Job listings have been saved to {csv_file_path}")

finally:
    driver.quit()  # Close the browser
```
